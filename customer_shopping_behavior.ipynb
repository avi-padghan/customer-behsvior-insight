```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns =df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.rename(columns={'purchase_amount_(usd)':'purchase_amount'},inplace=True)
```

```python
df.columns
```

```python
# create a columns of age group 
labels=['young_adult','adult','middel_aged','senior']
df['age_groups']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_groups']].sample(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create coulmn purchase_frequency_days
frequency_mapping={'Fortnightly':14, 'Weekly':7, 'Annually':365, 'Quarterly':90, 'Bi-Weekly':14,
       'Monthly':30, 'Every 3 Months':90}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].sample(10)
```
